# CSO Classifier Tester

```python
from cso_classifier import CSOClassifier
import json
from tqdm import tqdm
import datetime;
```

```python
with open("./sample_data_acm_cleaned.json","r") as fr:
    papers = json.load(fr)
```

```python
# cc = CSOClassifier(modules = "both", enhancement = "all", explanation = True, get_weights=True)
cc = CSOClassifier(modules = "both", enhancement = "all", explanation = True, get_weights=True, filter_by=["artificial intelligence"])
results = list()

with tqdm(total=len(papers)) as pbar:
    for paper in papers:
        t_paper = dict()
        t_paper["title"] = paper["title"]
        t_paper["abstract"] = paper["abstract"]
        t_paper["keywords"] = str(paper["keywords"])
        # print(pos, paper["link"])
        class_paper = cc.run(t_paper)
        results.append({**paper, **class_paper})
        pbar.update(1)
```

```python
results_file = f"./sample_data_acm_cleaned_classified{datetime.datetime.now()}.json"
with open(results_file, "w") as fw:
    json.dump(results, fw)
```

```python
for pos in range(len(results)):
    set1 = set(results[pos]["enhanced"]).union(results[pos]["union"])
    set2 = set(results[pos]["filtered_enhanced"]).union(results[pos]["filtered_union"])
    set3 = set1.intersection(set2)
    overlap = float(len(set3))/float(len(set1))
    print(f'For paper {pos:3}, the AI Overlap is {overlap:.2f} {"----- Core AI?" if overlap> 0.5 else ""}')
```
